# ***`KEY CHALLENGES IN E-COMMERCE :`***


*  ***`Financial performance:`***
      1. Analyzing total revenue growth over time.
      2. Identifying best-selling products and seasonal sales trends.
      3. Calculating the average order value to understand spending behavior.

*  ***`Customer Insights:`***
      1. `Identifying top-performing locations based on sales volume.`
      2. `Analyzing order cancellation trends and customer retention.`
      3. `Differentiating B2B vs. B2C customer behavior for targeted marketing`.

*  ***`Logistics and fulfillment optimization:`***
      1. `Addressing shipping delays by calculating average shipping times.`
      2. `Comparing orders fulfilled by Amazon vs. Merchant fulfillment.`
      3. `Improving order processing for faster deliveries.`


*   ***`Product management and Inventory optimization:`***
      1. `Identifying high-demand product categories.`
      2. `Analyzing the average quantity ordered for efficient stock management.`
      3. `Enhancing inventory control to prevent stockouts or overstocking.`

*  ***`Customer Satisfaction and returns:`***
      1. `Investigating high return and cancellation rates.`
      2. `Identifying problematic products/categories to enhance quality.`
      3. `Improving customer experience through better service policies`












***`Data Requirement:`***
      
  1. Order ID: A unique identifier assigned to each  order placed in the e-commerce system.
  2.  Date : Date on which product ordered.
  3. Size : Size of the product
  4. Quantity : The number of units of the product  ordered
  5. Status : Current status of the order (e.g., Shipped, Cancelled, Delivered to Buyer, etc.)
  6. currency : The currency used for the transaction.
  7. Amount : Amount of the transaction
  8. Fulfillment: Describes how the order was fulfilled(e.g., Amazon, Merchant)
  9. Sales Channel : The platform through which the sale was made(e.g., Amazon.in)
  10. ship-service-level : Shipping service level (e.g.,Standard, Expedited)
  11. Style : Style of the product
  12. SKU : Specific identifiers used for stock keeping and product design.
  13. ship-city, ship-state, ship-country: Details about the location to which the product was shipped.
  14. ship-postal-code: Postal code where the product was shipped.
  15. promotion_ids : Promotion IDs if any
  16. B2B : A boolean indicating whether the order is Business-to-Business (B2B)
  17. fulfilled_by : Who fulfilled the shipping (e.g., Easy Ship)
  18. Category : Category of the product

***`Removing Outlier `***

***`Product Demand Categorization`***

```python
# Calculate the total quantity sold for each product
product_demand = df.groupby('Category')['Qty'].count().sort_values(ascending=False).reset_index()

# Set thresholds to define high-demand and low-demand products (e.g., top 20% and bottom 20%)
high_demand_threshold = product_demand['Qty'].quantile(0.8)
low_demand_threshold = product_demand['Qty'].quantile(0.2)

 # Creating new column as "Demand_Category"
product_demand["Demand_Category"] = ' '

# Iterate through each row of the 'product_demand' DataFrame using iterrows()
for index,value in product_demand.iterrows():
  if value['Qty'] >= high_demand_threshold:
    product_demand.loc[index,"Demand_Category"] ='High Demand'
  elif value['Qty'] <= low_demand_threshold:
    product_demand.loc[index,"Demand_Category"] ='Low Demand'
  else:
    product_demand.loc[index,"Demand_Category"] ='Medium Demand'

display(product_demand)
```

***`Returned Rates of product Category Ordered`***

```python
# Calculate the total number of orders for each product category.
Total_orders_per_product=df.groupby('Category')['Order ID'].count()
Total_orders_per_product=Total_orders_per_product.sort_values(ascending=False).reset_index(name='Total_orders')

# Calculate the total number of returns for each product category.
Total_returns_per_product=Cancellation.groupby('Category')['Order ID'].count()
Total_returns_per_product=Total_returns_per_product.sort_values(ascending=False).reset_index(name='Total_returns')

# Merge the two DataFrames on the 'Category' column using an outer join.
merged_df = pd.merge(Total_orders_per_product, Total_returns_per_product,on='Category',how='outer')
# Fill any missing values in the 'Total_returns' and 'Total_orders' columns with 0.
merged_df['Total_returns'] =merged_df['Total_returns'].fillna(0)
merged_df['Total_orders'] =merged_df['Total_orders'].fillna(0)

# Calculate the return rate for each product category.
merged_df['Return_rate'] =(merged_df['Total_returns'] / merged_df['Total_orders']) * 100
merged_df['Return_rate'] = merged_df['Return_rate'].round(2)

# Display the merged DataFrame.
display(merged_df)
```

```python
%matplotlib inline
import plotly.express as px
import matplotlib.pyplot as plt

fig =px.bar(merged_df, x='Category', y='Return_rate',
            labels={'Return_rate': 'Return Rate (%)','Category': 'Product Category'},
            title='Return Rates by Product Category',
            color_discrete_sequence=['skyblue'],
            template='plotly_dark',
            text='Return_rate'
            )
fig.update_traces(textposition='outside',
                  marker=dict(line=dict(color='white',width=3))
                 )
fig.update_layout(
    xaxis=dict(tickangle=-45),
    font=dict(family='serif',size=15),
    title_x=0.5,
    showlegend=False
    )
fig.show()
```

***`Top 10 Product with Highest Returns`***

***` Top 10 Products with Highest Cancellation `***

```python
# Calculate the total quantity sold for each product (SKU)
qty_per_product=df.groupby('SKU')['Qty'].sum()
total_qty_per_product=qty_per_product.reset_index(name='Total_Qty')

# Calculate the total quantity of cancelled orders for each product (SKU).
total_cancel_qty_per_product=Cancellation.groupby('SKU')['Qty'].sum()
total_cancel_qty_per_product=total_cancel_qty_per_product.reset_index(name='Total_Cancellation')

# Merge the two DataFrames based on the 'SKU' column using an outer join
merge_cancel=pd.merge(total_qty_per_product,total_cancel_qty_per_product,on='SKU',how='outer')
merge_cancel['Total_Cancellation']=merge_cancel['Total_Cancellation'].fillna(0)
merge_cancel['Total_Qty']=merge_cancel['Total_Qty'].fillna(0)


# Get the top 10 products with the highest total cancellations.
Top_10_highest_cancellation=merge_cancel.sort_values(by='Total_Cancellation',ascending=False).head(10)

# Calculate the return rate for each of the top 10 products.
Top_10_highest_cancellation['Return_rate']=(Top_10_highest_cancellation['Total_Cancellation']/Top_10_highest_cancellation['Total_Qty'])*100
Top_10_highest_cancellation['Return_rate']=Top_10_highest_cancellation['Return_rate'].round(2)
# Display the DataFrame containing the top 10 products with the highest cancellations and their return rates.
display(Top_10_highest_cancellation)
```

***`Monthly Cancellation Rate `***

```python
# Convert the 'Date' column to datetime objects
df.loc[:,'Date']=pd.to_datetime(df['Date'])

# Extract the month from the 'Date' column and store it in a new 'Month' column
df.loc[:,'Month']=df['Date'].dt.to_period('M')

# Filter the DataFrame to include only cancelled orders
Cancel=df[(df['Status']=='Cancelled') |
 (df['Status']=='Shipped - Returned to Seller') |
 (df['Status']=='Shipped - Returning to Seller') |
 (df['Status']=='Shipped - Rejected by Buyer')
]

# Group the cancelled quantity  by month and sum the number of orders
montly_cancel_rate=Cancel.groupby('Month')['Qty'].sum().reset_index(name='Montly_Cancellation')
montly_cancel_rate=montly_cancel_rate.sort_values(by='Month',ascending=True)

# Group all quantity by month and sum the number of orders
montly_order_rate=df.groupby('Month')['Qty'].sum().reset_index(name='Montly_Orders')
montly_order_rate=montly_order_rate.sort_values(by='Month',ascending=True)

# Merge the two DataFrames on the 'Month' column
monthly_merge=pd.merge(montly_cancel_rate,montly_order_rate,on='Month',how='outer')

# Calculate the cancellation rate for each month
monthly_merge['Cancellation_rate']=(monthly_merge['Montly_Cancellation']/monthly_merge['Montly_Orders'])*100
monthly_merge['Cancellation_rate']=monthly_merge['Cancellation_rate'].round(2)

# Display the relevant columns of the merged DataFrame
display(monthly_merge[['Month','Montly_Orders','Montly_Cancellation','Cancellation_rate']])
```
